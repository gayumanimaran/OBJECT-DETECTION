```python
from imageai.Detection import ObjectDetection
```

```python
detector = ObjectDetection()

detector.setModelTypeAsRetinaNet()
```

```python
detector.setModelPath("/content/resnet50_coco_best_v2.1.0.h5")
```

```python
detector.loadModel()
```

```python
detection = detector.detectObjectsFromImage(input_image="/content/KinistinawPark.png", output_image_path="/content/objectdetection.jpg")
```
